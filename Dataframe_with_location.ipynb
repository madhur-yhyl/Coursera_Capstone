### Merging Latitute and Longitude Data with Wikipedia Data

```python
import pandas as pd 
df1 = pd.read_csv("Geospatial_Coordinates.csv") 
df1.head()
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re


def extract(s):
    mo = re.search(r'\((.*)\)', s)
    if mo:
        return mo.group(1)
    return ''

def split_string(string): 
  
    # Split the string based on space delimiter 
    list_string = string.split('/') 
      
    return list_string 

def join_string(list_string): 
  
    # Join the string based on '-' delimiter 
    string = ','.join(list_string) 
      
    return string 

col = 9
i = 0
post_code = []
borough = []
neighbourhood = []
with open('Wikipedia_file.html') as html_file:
    soup = BeautifulSoup(html_file, 'lxml')
    
table = soup.find('tbody')
tr = table.find_all('tr')
while i < 9:
    for row in tr:
        if (str(row.find_all('span')[i].text).lower() != 'not assigned'):
            post_code.append(str(row.find_all('b')[i].text))
            borough.append(str(row.find_all('span')[i].text).split('(')[0])
            neighbourhood.append(join_string(split_string(extract(str(row.find_all('span')[i].text)))))
    i+=1
cols = {'PostalCode':post_code, 'Borough':borough, 'Neighborhood':neighbourhood}
df2 = pd.DataFrame(cols)
df2
```

```python
result = pd.concat([df2, df1], axis=1, sort=False)
```
